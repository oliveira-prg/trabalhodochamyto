```python
import numpy as np
```

### <span style="color:pink">**Programa para operações com vetores**</span>

```python
print("(1)Soma\n(2)Subtração\n(3)Produto Interno\n(4)Norma\n(5)Distância Euclidiana\n")
operacao = int(input("Escolha a operação que você deseja realizar:"))

num_elementos = int(input("\nDigite o número de elementos a serem inseridos:"))
vetor1 = []
vetor2 = [] 

print("\nInsira os elementos do primeiro vetor:")
for i in range(num_elementos):
    valor1 = float(input(f"Elemento {i+1}: "))
    vetor1.append(valor1)

print("\nInsira os elementos do segundo vetor:")
for i in range(num_elementos):
    valor2 = float(input(f"Elemento {i+1}: "))
    vetor2.append(valor2)

match operacao:
    case 1:
        vetor_soma = np.add(vetor1, vetor2)
        print("A soma dos vetores é:", vetor_soma)
    case 2:
        vetor_sub = np.subtract(vetor1, vetor2)
        print("A subtração dos vetores é: ", vetor_sub)
    case 3:
        vetor_prod = np.inner(vetor1, vetor2)
        print("Produto interno:", vetor_prod)
    case 4:
        print("A norma do primeiro vetor é:", round(np.linalg.norm(vetor1), 2))
        print("A norma do segundo vetor é:", round(np.linalg.norm(vetor2), 2))
    case 5:
        diferenca = np.subtract(vetor1, vetor2)
        print("A distância entre os vetores é:", round(np.linalg.norm(diferenca), 3))
```

### <span style="color:pink">**Programa para operações com matrizes**</span>

```python
print("(1)Soma\n(2)Subtração\n(3)Multiplição\n(4)Determinante\n(5)Inversa\n")
quant = int(input("Quantas matrizes você quer trabalhar?(máximo duas!)"))
operacao = int(input("Escolha a operação que você deseja realizar:"))

linhas = int(input("\nDigite o número de linhas a serem inseridos:"))
colunas = int(input("\nDigite o número de colunas a serem inseridos:"))

matriz1 = []
matriz2 = []


match quant:
    case 1:
        for i in range(linhas):
            linha = []
            for j in range(colunas):
                valor = float(input(f"Digite o valor para [{i}][{j}]: "))
                linha.append(valor)
            matriz1.append(linha)
            
        matriz1 = np.array(matriz1)

        print("\nMatriz:")
        for linha in matriz1:
            print(linha)
    case 2:
        for i in range(linhas):
            linha = []
            for j in range(colunas):
                valor = float(input(f"Digite o valor para [{i}][{j}]: "))
                linha.append(valor)
            matriz1.append(linha)
                
        for i in range(linhas):
            linha = []
            for j in range(colunas):
                valor = float(input(f"Digite o valor para [{i}][{j}]: "))
                linha.append(valor)
            matriz2.append(linha)
            
        matriz1 = np.array(matriz1)
        matriz2 = np.array(matriz2)

        print("\nMatriz:")
        for linha in matriz1:
            print(linha)
            
        print("\nMatriz:")
        for linha in matriz2:
            print(linha)
    case _:
        print("ERRO: quantidade inválida de matrizes!")
        exit()


match operacao:
    case 1:
        matriz_soma = np.add(matriz1, matriz2)
        print("A soma das matrizes é:", matriz_soma)
    case 2:
        matriz_sub = np.subtract(matriz1, matriz2)
        print("A diferença das matrizes é: ", matriz_sub)
    case 3:
        matriz_prod = matriz1 @ matriz2
        print("Produto das matrizes:", matriz_prod)
    case 4:
        print("A a determinante da primeira matriz é:", np.round(np.linalg.det(matriz1), 2))
        if quant == 2:
            print("A a determinante da segunda é:", np.round(np.linalg.det(matriz2), 2))
    case 5:
        try:
            print("A inversa da primeira é:", np.round(np.linalg.inv(matriz1), 3))
            if quant == 2:
                print("A inversa da segunda é:", np.round(np.linalg.inv(matriz2), 3))
        except np.linalg.LinAlgError:
            print("Uma das matrizes não é invertível (determinante = 0).")
            
    case _:
        print("ERRO operação inválida")
```
